```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
file_path = '/content/A1602_SAM03_TB_DC_00_2021_C03_F_GR.csv'
```

```python
df = pd.read_csv(
    file_path,
    skiprows=[1,2,4,5],
    header=1
)
```

```python
# pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df.tail()
```

```python
columns_to_drop = [4,5,6,7]
df = df.drop(df.columns[columns_to_drop], axis=1)
```

```python
df.dtypes
```

```python
df.columns
```

```python
df.rename(columns={
    "Γεωγρφικό επίπεδο ": "geo_level",
    "Γεωγραφικός Κωδικός": "geo_code",
    "Περιγραφή/Ομάδες ηλικιών": "age_group",
    "Σύνολο": "total"
    }, inplace = True)
```

```python
df['total']=df['total'].str.replace(',','')
```

```python
df['total']
```

```python
df['total'] = pd.to_numeric(df['total'])
```

```python
df_averages = (
    df.groupby('geo_code')['total']
      .apply(lambda x: round(x.iloc[2] / x.iloc[0], 3))
      .reset_index(name='average')
)
```

```python
df_averages
```

```python
df_averages.tail()
```

```python
df_averages['geo_code'] = (
    df_averages['geo_code']
        .astype(str)
        .str.replace(r'\.0$', '', regex=True)  # remove trailing .0
        .str[3:]
)
```

```python
df_averages.tail()
```

```python
df_averages['geo_code'] = df_averages['geo_code'].astype(str)
```

```python
df_averages = df_averages[df_averages['geo_code'].str.len() == 8]
```

```python
df_averages.info()
```

```python
df_averages.describe()
```

```python
fig = px.box(df_averages, y="average")
fig.show()
```
